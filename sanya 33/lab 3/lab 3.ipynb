```python
import matplotlib.pyplot as plt

def bresenham_line (x1, y1, x2, y2):
    xes, yes = [], []

    dx = abs(x2-x1)
    dy = abs(y2-y1)

    sx = 1 if x2>=x1 else -1
    sy = 1 if y2>=y1 else -1 

    x,y = x1, y1

    if dx >= dy:
        p = 2*dy-dx
        for _ in range (dx+1):
            xes.append(x)
            yes.append(y)
            y += sy
            if p>= 0:
                x += sx
                p += 2*dx - 2*dy
            else:
                p += 2*dx
    return xes, yes

def plot_bresenham (x1 , y1 , x2 , y2): 
    xes, yes = bresenham_line (x1, y1, x2, y2)
    plt.figure(figsize =(6, 6))
    plt.plot(xes, yes , marker='o', linestyle ='-', color='purple')
    plt.title("Bresenham Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show ()

plot_bresenham (2, 3, 10, 8)
```

```python
import matplotlib.pyplot as plt

def bresenham_line (x1, y1, x2, y2):
    xes, yes = [], []

    dx = abs(x2-x1)
    dy = abs(y2-y1)

    sx = 1 if x2>=x1 else -1
    sy = 1 if y2>=y1 else -1 

    x,y = x1, y1

    if dx >= dy:
        p = 2*dy-dx
        for _ in range (dx+1):
            xes.append(x)
            yes.append(y)
            y += sy
            if p>= 0:
                x += sx
                p += 2*dx - 2*dy
            else:
                p += 2*dx
    return xes, yes
    
def dda(x1, y1, x2, y2):
    x_pts, y_pts = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1

    for i in range(steps + 1):
        x_pts.append(round(x))
        y_pts.append(round(y))
        x += x_inc
        y += y_inc

    return x_pts, y_pts

x1, y1, x2, y2 = 2, 3, -10, 15

x_d, y_d = dda(x1, y1, x2, y2)
x_b, y_b = bresenham_line(x1, y1, x2, y2)

plt.plot(x_d, y_d, marker='o', label='DDA')
plt.plot(x_b, y_b, marker='x', label='Bresenham')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.title("DDA vs Bresenham (Different Octant)")
plt.show()
```
