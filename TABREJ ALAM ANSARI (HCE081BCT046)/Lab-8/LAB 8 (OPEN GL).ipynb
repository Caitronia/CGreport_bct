```python
from OpenGL.GL import *
from OpenGL.GLUT import *

def display():
    glClear(GL_COLOR_BUFFER_BIT)

    # Point
    glColor3f(1, 0, 0)
    glPointSize(6)
    glBegin(GL_POINTS)
    glVertex2f(-0.7, 0.7)
    glEnd()

    # Line
    glColor3f(0, 1, 0)
    glBegin(GL_LINES)
    glVertex2f(-0.7, 0.3)
    glVertex2f(-0.2, 0.3)
    glEnd()

    # Triangle
    glColor3f(0, 0, 1)
    glBegin(GL_TRIANGLES)
    glVertex2f(0.2, 0.6)
    glVertex2f(0.6, 0.6)
    glVertex2f(0.4, 0.2)
    glEnd()

    # Polygon
    glColor3f(1, 1, 0)
    glBegin(GL_POLYGON)
    glVertex2f(-0.2, -0.2)
    glVertex2f(0.2, -0.2)
    glVertex2f(0.3, -0.5)
    glVertex2f(0.0, -0.7)
    glVertex2f(-0.3, -0.5)
    glEnd()

    glutSwapBuffers()   # FAST display

def main():
    glutInit()
    glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB)
    glutInitWindowSize(500, 500)
    glutCreateWindow(b"OpenGL Basic Shapes")

    glClearColor(1, 1, 1, 1)
    glutDisplayFunc(display)
    glutMainLoop()

main()
```
