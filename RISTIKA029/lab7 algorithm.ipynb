

### **Algorithm**

1. **Start**

2. **Input the 3D Object**

   * Read number of vertices ( n )
   * Input coordinates of each vertex
     [
     P(x, y, z)
     ]

3. **Convert to Homogeneous Coordinates**

   * Represent each point as
     [
     P(x, y, z, 1)
     ]

4. **Select Geometric Transformation**

   * Translation
   * Scaling
   * Rotation (about X, Y, or Z axis)
   * Or combination of transformations

5. **Form Transformation Matrix**

   * Construct required matrix (T / S / R)
   * If multiple transformations are required, compute composite matrix:
     [
     C = T \times R \times S
     ]

6. **Apply Transformation**

   * For each vertex:
     [
     P' = C \times P
     ]
   * Store transformed coordinates ( (x', y', z') )

7. **Select Projection Type (Visualization Stage)**

   * Parallel Projection
   * Perspective Projection

8. **Apply Projection**

   **(a) Parallel Projection**
   [
   x'' = x'
   ]
   [
   y'' = y'
   ]

   **(b) Perspective Projection**
   [
   x'' = \frac{x'}{1 + \frac{z'}{d}}
   ]
   [
   y'' = \frac{y'}{1 + \frac{z'}{d}}
   ]

   Where ( d ) = distance of viewer from projection plane

9. **Viewport Transformation**

   * Map projected coordinates to screen coordinates

   [
   x_s = x_{min} + (x'' - x_{wmin}) \frac{(x_{max} - x_{min})}{(x_{wmax} - x_{wmin})}
   ]

   [
   y_s = y_{min} + (y'' - y_{wmin}) \frac{(y_{max} - y_{min})}{(y_{wmax} - y_{wmin})}
   ]

10. **Display the Object**

    * Draw lines between projected vertices
    * Render final image on screen

11. **Stop**
