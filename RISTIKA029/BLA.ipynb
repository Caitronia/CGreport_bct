```python
import matplotlib.pyplot as plt

# Input points
x1, y1 = 2, 2
x2, y2 = 10, 6

dx = x2 - x1
dy = y2 - y1

p = 2 * dy - dx

x = x1
y = y1

x_points = []
y_points = []

x_points.append(x)
y_points.append(y)

while x < x2:
    x += 1
    if p < 0:
        p = p + 2 * dy
    else:
        y += 1
        p = p + 2 * dy - 2 * dx
    x_points.append(x)
    y_points.append(y)

plt.plot(x_points, y_points, marker='o')
plt.title("Bresenham Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

Alogrithm
Step 1: Start
Step 2: Input starting point (x1,y1) and ending point (x2,y2)
Step 3: Compute
   dx = x2 − x1
   
   dy = y2 − y1
Step 4: Initialize decision parameter
         
         p = 2dy − dx
Step 5: Initialize
  x = x1
  
  y = y1  
Step 6: Plot the initial point (x, y)
Step 7: While x < x2, repeat
         If p < 0

         x = x + 1

         p = p + 2dy

        Else

            x = x + 1
            y = y + 1
            p = p + 2dy − 2dx
Step 9: Plot point (x,y).
Step 8: Stop
