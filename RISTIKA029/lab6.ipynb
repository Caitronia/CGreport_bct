## 1.Translate a triangle with a translation factor tx and ty

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([
    [0, 0, 1],
    [2, 0, 1],
    [1, 2, 1]
])

tx = 3
ty = 2

T = np.array([
    [1, 0, tx],
    [0, 1, ty],
    [0, 0, 1]
])
translated_triangle = (T @ triangle.T).T
orig_xy = triangle[:, :2]
trans_xy = translated_triangle[:, :2]
orig_plot = np.vstack((orig_xy, orig_xy[0]))
trans_plot = np.vstack((trans_xy, trans_xy[0]))
plt.plot(orig_plot[:,0], orig_plot[:,1], 'b-')
plt.plot(trans_plot[:,0], trans_plot[:,1], 'r--')
plt.scatter(orig_xy[:,0], orig_xy[:,1], color='blue')
plt.scatter(trans_xy[:,0], trans_xy[:,1], color='red')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```

## 2.Scale a rectangle about a origin

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 6, 6, 1], 
    [1, 1, 8, 8],  
    [1, 1, 1, 1]    
])

# scaling factors
sx = 2
sy = 2

# scaling matrix (about origin)
scaling_mat = np.array([
    [sx, 0,  0],
    [0,  sy, 0],
    [0,  0,  1]
])

# scaling
scaled_rect = scaling_mat @ rectangle

# plot
plt.figure(figsize=(5,5))

# original rectangle
plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

# scaled rectangle
plt.plot(
    [scaled_rect[0,0], scaled_rect[0,1], scaled_rect[0,2], scaled_rect[0,3], scaled_rect[0,0]],
    [scaled_rect[1,0], scaled_rect[1,1], scaled_rect[1,2], scaled_rect[1,3], scaled_rect[1,0]],
    color='forestgreen', linestyle='--', label='scaled rectangle'
)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scaling of a rectangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

## 3. Rotate a triangle about origin.

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[0, 8, 1],
                     [0, 0, 2],
                     [1, 1, 1]])

theta = 45
rad = np.deg2rad(theta)

rotation_mat = np.array([
    [np.cos(rad), -np.sin(rad), 0],
    [np.sin(rad),  np.cos(rad), 0],
    [0, 0, 1]
])

rotated_tri = rotation_mat @ triangle

plt.figure()

plt.plot([triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
         [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
         label="Original")

plt.plot([rotated_tri[0,0], rotated_tri[0,1], rotated_tri[0,2], rotated_tri[0,0]],
         [rotated_tri[1,0], rotated_tri[1,1], rotated_tri[1,2], rotated_tri[1,0]],
         linestyle="--", label="Rotated")

plt.legend()
plt.axis("equal")
plt.grid()
plt.show()
```

## 4. Reflect a triangle
a. about x-axis

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 9, 1],   
    [0, 0, 45],  
    [1, 1, 1]   
])

#reflection about x axis
ref=np.array([
    [1, 0, 0],
    [0, -1, 0],
    [0, 0, 1]
])

reflected_tri = ref @ triangle

#plot
plt.figure(figsize=(5,5))

#original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='black', linestyle='-', label = 'original triangle'
)

#translated triangle
plt.plot(
    [reflected_tri[0,0], reflected_tri[0,1], reflected_tri[0,2], reflected_tri[0,0]],
    [reflected_tri[1,0], reflected_tri[1,1], reflected_tri[1,2], reflected_tri[1,0]],
    color='violet', label = 'reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about x axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

d. about y=mx+c

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([
    [1, 1, 1],
    [4, 1, 1],
    [2, 4, 1]
])

m = 1
c = 1
theta = np.arctan(m)

T1 = np.array([[1,0,0],[0,1,-c],[0,0,1]])
R1 = np.array([[np.cos(-theta),-np.sin(-theta),0],
               [np.sin(-theta), np.cos(-theta),0],
               [0,0,1]])

Rx = np.array([[1,0,0],[0,-1,0],[0,0,1]])

R2 = np.array([[np.cos(theta),-np.sin(theta),0],
               [np.sin(theta), np.cos(theta),0],
               [0,0,1]])

T2 = np.array([[1,0,0],[0,1,c],[0,0,1]])

final_matrix = T2 @ R2 @ Rx @ R1 @ T1

reflected = (final_matrix @ triangle.T).T
t1 = np.vstack((triangle[:, :2], triangle[0, :2]))
t2 = np.vstack((reflected[:, :2], reflected[0, :2]))

plt.plot(t1[:,0], t1[:,1], 'b-', label="Original")
plt.plot(t2[:,0], t2[:,1], 'r--', label="Reflected")
plt.gca().set_aspect('equal')
plt.grid()
plt.legend()
plt.title("Reflection about y = mx + c")
plt.show()
```

## 5. Shear a rectangle
a. about x direction

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 8, 8, 1],
    [1, 1, 2, 2],
    [1, 1, 1, 1]
])

# shear factor in x direction
shx = 1.5

# x-direction shear matrix
shear_x = np.array([
    [1, shx, 0],
    [0, 1,   0],
    [0, 0,   1]
])

# shearing
sheared_rect = shear_x @ rectangle

# plot
plt.figure(figsize=(5,5))

# original rectangle
plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

# sheared rectangle
plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='olivedrab', linestyle='--', label='x-direction shear'
)
plt.title("shearing of rectangle in x direction")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 7, 7, 1],
    [1, 1, 2, 2],
    [1, 1, 1, 1]
])

# shear factors
shx = 1.0
shy = 0.7

# shearing in both directions
shear_xy = np.array([
    [1,   shx, 0],
    [shy, 1,   0],
    [0,   0,   1]
])

# shearing
sheared_rect = shear_xy @ rectangle

# plot
plt.figure(figsize=(5,5))

plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='dimgray', linestyle='--', label='shearing on both directions'
)

plt.title("shearing of rectangle in both directions")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```
