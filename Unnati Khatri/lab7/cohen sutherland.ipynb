Given a clipping window ,draw a line which lies:
1. Completely inside
2. Completely outside
3. Partially inside (any two conditions)
4. Partially inside intersecting two points

Solution:
let the coordinates of windows be 
xmin = 2 , ymin = 2
xmax = 8 , ymax = 8


```python
import matplotlib.pyplot as plt

xmin, ymin = 2, 2
xmax, ymax = 8, 8

# Draw clipping window
plt.plot([xmin, xmax, xmax, xmin, xmin],
         [ymin, ymin, ymax, ymax, ymin], 'k', linewidth=2)

# Line completely inside (accepted)
plt.plot([3, 7], [4, 6], 'b-', linewidth=2)

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.title("Line Completely Inside")
plt.show()
```

```python
import matplotlib.pyplot as plt

xmin, ymin = 2, 2
xmax, ymax = 8, 8

# Draw clipping window
plt.plot([xmin, xmax, xmax, xmin, xmin],
         [ymin, ymin, ymax, ymax, ymin], 'k', linewidth=2)

# Rejected part (outside)
plt.plot([1, 2], [5, 5], 'r--', linewidth=2)

# Accepted part (inside)
plt.plot([2, 5], [5, 5], 'g-', linewidth=2)

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.title("Partially Inside (One Intersection)")
plt.show()
```

```python
import matplotlib.pyplot as plt

xmin, ymin = 2, 2
xmax, ymax = 8, 8

# Draw clipping window
plt.plot([xmin, xmax, xmax, xmin, xmin],
         [ymin, ymin, ymax, ymax, ymin], 'k', linewidth=2)

# Rejected parts (outside)
plt.plot([1, 2], [6, 6], 'r--', linewidth=2)
plt.plot([8, 9], [6, 6], 'r--', linewidth=2)

# Accepted part (inside)
plt.plot([2, 8], [6, 6], 'b-', linewidth=2)

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid(True)
plt.title("Partially Inside (Two Intersections)")
plt.show()
```
