# LAB-2: BLA(Bresenham's Line Drawing Algorithm)

## Algorithm

**Step 1:** Start  

**Step 2:** Input starting point (x1, y1) and ending point (x2, y2)

**Step 3:** Compute  
dx = x2 − x1  
dy = y2 − y1  

**Step 4:** Initialize decision parameter  
p = 2dy − dx  

**Step 5:** Initialize  
x = x1  
y = y1  

**Step 6:** Plot the initial point (x, y)

**Step 7:** While x < x2, repeat  

If p < 0  
x = x + 1  
p = p + 2dy  

Else  
x = x + 1  
y = y + 1  
p = p + 2dy − 2dx  

**Step 8:** Plot point (x, y)

**Step 9:** Stop


```python
import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    xes, yes = [], []
    
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    sx = 1 if x2 >= x1 else -1
    sy = 1 if y2 >= y1 else -1
    
    x, y = x1, y1

    if dx >= dy:
        p = 2 * dy - dx
        for _ in range(dx + 1):
            xes.append(x)
            yes.append(y)
            x += sx
            if p >= 0:
                y += sy
                p += 2 * dy - 2 * dx
            else:
                p += 2 * dy
    else:
        p = 2 * dx - dy
        for _ in range(dy + 1):
            xes.append(x)
            yes.append(y)
            y += sy
            if p >= 0:
                x += sx
                p += 2 * dx - 2 * dy
            else:
                p += 2 * dx

    return xes, yes

def plot_bresenham(x1, y1, x2, y2):
    xes, yes = bresenham_line(x1, y1, x2, y2)
    
    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker='o', linestyle='-')
    plt.title("Bresenham Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()


```
