# LAB-1: DDA(Digital Differential Analyzer)

## Algorithm

**Step 1:** Start  

**Step 2:** Input starting point (x1, y1) and ending point (x2, y2)

**Step 3:** Compute  
dx = x2 − x1  
dy = y2 − y1  

**Step 4:** Calculate number of steps  
steps = max(|dx|, |dy|)

**Step 5:** Compute increment values  
x_inc = dx / steps  
y_inc = dy / steps  

**Step 6:** Initialize  
x = x1  
y = y1  

**Step 7:** Plot the initial point (x, y)

**Step 8:** Repeat for steps times  
x = x + x_inc  
y = y + y_inc  

**Step 9:** Plot point (round(x), round(y))

**Step 10:** Stop


```python
import matplotlib.pyplot as plt

def draw_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1
    xs = [round(x)]
    ys = [round(y)]

    for _ in range(steps):
        x += x_inc
        y += y_inc
        xs.append(round(x))
        ys.append(round(y))

    print("X points:", xs)
    print("Y points:", ys)

    plt.plot(xs, ys)
    plt.show()


x1 = int(input("x1: "))
y1 = int(input("y1: "))
x2 = int(input("x2: "))
y2 = int(input("y2: "))

draw_line(x1, y1, x2, y2)
```
