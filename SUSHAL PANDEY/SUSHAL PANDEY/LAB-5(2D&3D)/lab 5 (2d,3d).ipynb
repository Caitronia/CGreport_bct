**1. Translate a triangle with translation factors tx and ty.**

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([
    [0, 0, 1],
    [2, 0, 1],
    [1, 1, 1]
])

#translation factor
tx = 0.5
ty = 2

#translation matrix
translation_mat = np.array([
    [1, 0, tx],
    [0, 1, ty],
    [0, 0, 1]
])

#translation
translated_tri = translation_mat @ triangle

#plot
plt.figure(figsize=(5,5))

#original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color = 'black', label = 'original triangle'
)

#translated triangle
plt.plot(
    [translated_tri[0,0], translated_tri[0,1], translated_tri[0,2], translated_tri[0,0]],
    [translated_tri[1,0], translated_tri[1,1], translated_tri[1,2], translated_tri[1,0]],
    color = 'Blue', linestyle='--', label = 'translated triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("translation of a triangle")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

**2. Scale a rectangle about an origin.**

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 4, 4, 1], 
    [1, 1, 3, 3],  
    [1, 1, 1, 1]    
])

# scaling factors
sx = 2
sy = 2

# scaling matrix (about origin)
scaling_mat = np.array([
    [sx, 0,  0],
    [0,  sy, 0],
    [0,  0,  1]
])

# scaling
scaled_rect = scaling_mat @ rectangle

# plot
plt.figure(figsize=(5,5))

# original rectangle
plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='green', label='original rectangle'
)

# scaled rectangle
plt.plot(
    [scaled_rect[0,0], scaled_rect[0,1], scaled_rect[0,2], scaled_rect[0,3], scaled_rect[0,0]],
    [scaled_rect[1,0], scaled_rect[1,1], scaled_rect[1,2], scaled_rect[1,3], scaled_rect[1,0]],
    color='crimson', linestyle='--', label='scaled rectangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scaling of a rectangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

**3. Rotate a triangle about an origin.**

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

# rotation angle (degrees)
theta = 45

# convert degrees to radians
rad = np.deg2rad(theta)

# rotation matrix (about origin)
rotation_mat = np.array([
    [np.cos(rad), -np.sin(rad), 0],
    [np.sin(rad),  np.cos(rad), 0],
    [0,            0,           1]
])

# rotation
rotated_tri = rotation_mat @ triangle

# plot
plt.figure(figsize=(5,5))

# original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='blue', label='original triangle'
)

# rotated triangle
plt.plot(
    [rotated_tri[0,0], rotated_tri[0,1], rotated_tri[0,2], rotated_tri[0,0]],
    [rotated_tri[1,0], rotated_tri[1,1], rotated_tri[1,2], rotated_tri[1,0]],
    color = 'red', linestyle='--', label='rotated triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("rotation of a triangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

**4. Reflect a triangle**

**a. about x-axis**

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

#reflection about x axis
ref=np.array([
    [1, 0, 0],
    [0, -1, 0],
    [0, 0, 1]
])

reflected_tri = ref @ triangle

#plot
plt.figure(figsize=(5,5))

#original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='blue', linestyle='-', label = 'original triangle'
)

#translated triangle
plt.plot(
    [reflected_tri[0,0], reflected_tri[0,1], reflected_tri[0,2], reflected_tri[0,0]],
    [reflected_tri[1,0], reflected_tri[1,1], reflected_tri[1,2], reflected_tri[1,0]],
    color='black', label = 'reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about x axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

**d. about y = mx + c**

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

m = 1
c = 1

theta = np.arctan(m)

# translation matrices
T1 = np.array([
    [1, 0, 0],
    [0, 1, -c],
    [0, 0, 1]
])

T2 = np.array([
    [1, 0, 0],
    [0, 1, c],
    [0, 0, 1]
])

# rotation matrices
R1 = np.array([
    [ np.cos(theta),  np.sin(theta), 0],
    [-np.sin(theta),  np.cos(theta), 0],
    [ 0,              0,             1]
])

R2 = np.array([
    [ np.cos(theta), -np.sin(theta), 0],
    [ np.sin(theta),  np.cos(theta), 0],
    [ 0,              0,             1]
])

# reflection about x-axis
Rx = np.array([
    [1,  0, 0],
    [0, -1, 0],
    [0,  0, 1]
])

# combined reflection matrix
reflection_mat = T2 @ R2 @ Rx @ R1 @ T1

# reflection
reflected_tri = reflection_mat @ triangle

#plot
plt.figure(figsize=(6,6))

#original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='yellow', linestyle='-', label = 'original triangle'
)

#translated triangle
plt.plot(
    [reflected_tri[0,0], reflected_tri[0,1], reflected_tri[0,2], reflected_tri[0,0]],
    [reflected_tri[1,0], reflected_tri[1,1], reflected_tri[1,2], reflected_tri[1,0]],
    color='green', linestyle='--', label = 'reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about y = mx + c")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

**5. Shear a rectangle**

**a. about x direction**

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 4, 4, 1],
    [1, 1, 3, 3],
    [1, 1, 1, 1]
])

# shear factor in x direction
shx = 1.5

# x-direction shear matrix
shear_x = np.array([
    [1, shx, 0],
    [0, 1,   0],
    [0, 0,   1]
])

# shearing
sheared_rect = shear_x @ rectangle

# plot
plt.figure(figsize=(5,5))

# original rectangle
plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='green', label='original rectangle'
)

# sheared rectangle
plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='olivedrab', linestyle='--', label='x-direction shear'
)

plt.title("shearing of rectangle in x-direction")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```
