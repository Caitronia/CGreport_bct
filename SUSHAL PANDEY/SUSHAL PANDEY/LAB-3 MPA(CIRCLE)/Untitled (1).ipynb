```python
import matplotlib.pyplot as plt


def plot_circle_points(xc, yc, x, y, x_points, y_points):
    points = [
        ( x + xc,  y + yc),
        (-x + xc,  y + yc),
        ( x + xc, -y + yc),
        (-x + xc, -y + yc),
        ( y + xc,  x + yc),
        (-y + xc,  x + yc),
        ( y + xc, -x + yc),
        (-y + xc, -x + yc),
    ]

    for px, py in points:
        x_points.append(px)
        y_points.append(py)


def midpoint_circle(radius, xc=0, yc=0):
    
    x = 0
    y = radius
    p = 1 - radius

    x_points, y_points = [], []

    plot_circle_points(xc, yc, x, y, x_points, y_points)

    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

        plot_circle_points(xc, yc, x, y, x_points, y_points)

    return x_points, y_points
def plot_midpoint_circle(r,xc=0,yc=0):
    xes,yes =midpoint_circle(r,xc,yc)
    plt.figure(figsize=(6,6))
    plt.scatter(xes,yes,marker='.',color='red')
    plt.title("midpoint")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()
    
              
```

```python
plot_midpoint_circle(20,5,6)
```

```python
plot_midpoint_circle(5,5,6)
```

```python
plot_midpoint_circle(60,-5,-6)
```

```python
plot_midpoint_circle(20,20,-6)
```
