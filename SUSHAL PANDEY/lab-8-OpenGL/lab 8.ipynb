Lab Task
Draw fundamental shapes like points, lines, triangles, and polygons

```python
#point

import glfw
from OpenGL.GL import *

# Initialize GLFW
if not glfw.init():
    raise Exception("GLFW not initialized")

# Create window
window = glfw.create_window(600, 600, "OpenGL Basics", None, None)
glfw.make_context_current(window)

# Render loop
while not glfw.window_should_close(window):
    glClear(GL_COLOR_BUFFER_BIT)

    glPointSize(10)
    glBegin(GL_POINTS)
    glVertex2f(0.0, 0.0)
    glEnd()

    glfw.swap_buffers(window)   # Render
    glfw.poll_events()          # Check events

glfw.terminate()
```

```python
#line

import glfw
from OpenGL.GL import *

# Initialize GLFW
if not glfw.init():
    raise Exception("GLFW not initialized")

window = glfw.create_window(600, 600, "Line", None, None)
glfw.make_context_current(window)

while not glfw.window_should_close(window):
    glClear(GL_COLOR_BUFFER_BIT)

    glBegin(GL_LINES)
    glVertex2f(-0.5, 0.0)  
    glVertex2f(0.5, 0.0)   
    glEnd()

    glfw.swap_buffers(window)
    glfw.poll_events()

glfw.terminate()
```

```python
#triangle

import glfw
from OpenGL.GL import *

if not glfw.init():
    raise Exception("GLFW not initialized")

window = glfw.create_window(600, 600, "Triangle", None, None)
glfw.make_context_current(window)

while not glfw.window_should_close(window):
    glClear(GL_COLOR_BUFFER_BIT)

    glBegin(GL_TRIANGLES)
    glVertex2f(0.0, 0.5)    
    glVertex2f(-0.5, -0.5)   
    glVertex2f(0.5, -0.5)    
    glEnd()

    glfw.swap_buffers(window)
    glfw.poll_events()

glfw.terminate()
```

```python
#polygon

import glfw
from OpenGL.GL import *

if not glfw.init():
    raise Exception("GLFW not initialized")

window = glfw.create_window(600, 600, "Polygon", None, None)
glfw.make_context_current(window)

while not glfw.window_should_close(window):
    glClear(GL_COLOR_BUFFER_BIT)

    glBegin(GL_POLYGON)
    glVertex2f(0.0, 0.6)
    glVertex2f(-0.5, 0.2)
    glVertex2f(-0.3, -0.5)
    glVertex2f(0.3, -0.5)
    glVertex2f(0.5, 0.2)
    glEnd()

    glfw.swap_buffers(window)
    glfw.poll_events()

glfw.terminate()
```
