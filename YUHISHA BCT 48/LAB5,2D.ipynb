Translate a triangle with translation factors tx and ty

```python
import matplotlib.pyplot as plt

# Original triangle vertices
A = (0, 0)
B = (2, 0)
C = (1, 3)

# Translation factors
tx, ty = 4, 2

# Apply translation
A_prime = (A[0] + tx, A[1] + ty)
B_prime = (B[0] + tx, B[1] + ty)
C_prime = (C[0] + tx, C[1] + ty)

# Plot original triangle
x_orig = [A[0], B[0], C[0], A[0]]
y_orig = [A[1], B[1], C[1], A[1]]
plt.plot(x_orig, y_orig, 'b-', label='Original Triangle')

# Plot translated triangle
x_trans = [A_prime[0], B_prime[0], C_prime[0], A_prime[0]]
y_trans = [A_prime[1], B_prime[1], C_prime[1], A_prime[1]]
plt.plot(x_trans, y_trans, 'r--', label='Translated Triangle')

# Add labels and legend
plt.scatter(*zip(A, B, C), color='blue')
plt.scatter(*zip(A_prime, B_prime, C_prime), color='red')
plt.legend()
plt.axis('equal')
plt.title("Triangle Translation")
plt.show()
```

Scale a rectangle about a origin

```python
import matplotlib.pyplot as plt

# Original rectangle vertices (counter-clockwise)
# Let's say rectangle corners are at (1,1), (3,1), (3,2), (1,2)
rectangle = [(1,1), (3,1), (3,2), (1,2)]

# Scaling factors
sx, sy = 2, 1.5   # scale x by 2, y by 1.5

# Scale about origin (0,0)
scaled_rectangle = [(x * sx, y * sy) for (x,y) in rectangle]

# Close the shapes for plotting
x_orig = [p[0] for p in rectangle] + [rectangle[0][0]]
y_orig = [p[1] for p in rectangle] + [rectangle[0][1]]

x_scaled = [p[0] for p in scaled_rectangle] + [scaled_rectangle[0][0]]
y_scaled = [p[1] for p in scaled_rectangle] + [scaled_rectangle[0][1]]

# Plot both rectangles
plt.plot(x_orig, y_orig, 'b-', label='Original Rectangle')
plt.plot(x_scaled, y_scaled, 'r--', label='Scaled Rectangle')

# Mark vertices
plt.scatter(*zip(*rectangle), color='blue')
plt.scatter(*zip(*scaled_rectangle), color='red')

plt.legend()
plt.axis('equal')
plt.title("Scaling a Rectangle about Origin")
plt.show()
```

Rotate a triangle about a origin

```python
import matplotlib.pyplot as plt
import math

# Original triangle vertices
A = (0, 0)
B = (2, 0)
C = (1, 3)

# Rotation angle in degrees
theta = 45
theta_rad = math.radians(theta)

# Rotation function
def rotate_point(point, angle_rad):
    x, y = point
    x_new = x * math.cos(angle_rad) - y * math.sin(angle_rad)
    y_new = x * math.sin(angle_rad) + y * math.cos(angle_rad)
    return (x_new, y_new)

# Rotate all vertices
A_rot = rotate_point(A, theta_rad)
B_rot = rotate_point(B, theta_rad)
C_rot = rotate_point(C, theta_rad)

# Plot original triangle
x_orig = [A[0], B[0], C[0], A[0]]
y_orig = [A[1], B[1], C[1], A[1]]
plt.plot(x_orig, y_orig, 'b-', label='Original Triangle')

# Plot rotated triangle
x_rot = [A_rot[0], B_rot[0], C_rot[0], A_rot[0]]
y_rot = [A_rot[1], B_rot[1], C_rot[1], A_rot[1]]
plt.plot(x_rot, y_rot, 'r--', label=f'Rotated {theta}')

# Mark vertices
plt.scatter(*zip(A, B, C), color='blue')
plt.scatter(*zip(A_rot, B_rot, C_rot), color='red')

plt.legend()
plt.axis('equal')
plt.title("Triangle Rotation about Origin")
plt.show()
```

Reflection about x-axis

```python
import matplotlib.pyplot as plt

# Original triangle vertices
triangle = [(0,0), (2,0), (1,3)]

# Reflect across x-axis
reflected_triangle = [(x, -y) for (x,y) in triangle]

# Close shapes for plotting
x_orig = [p[0] for p in triangle] + [triangle[0][0]]
y_orig = [p[1] for p in triangle] + [triangle[0][1]]

x_ref = [p[0] for p in reflected_triangle] + [reflected_triangle[0][0]]
y_ref = [p[1] for p in reflected_triangle] + [reflected_triangle[0][1]]

# Plot
plt.plot(x_orig, y_orig, 'b-', label='Original Triangle')
plt.plot(x_ref, y_ref, 'r--', label='Reflected across X-axis')

plt.scatter(*zip(*triangle), color='blue')
plt.scatter(*zip(*reflected_triangle), color='red')

plt.axhline(0)   # x-axis
plt.legend()
plt.axis('equal')
plt.title("Triangle Reflection across X-axis")
plt.show()
```

Reflection about y=mx+c

```python
import matplotlib.pyplot as plt

def reflect_point(x0, y0, m, c):
    # Line in form ax + by + c = 0
    a, b = m, -1
    # Formula for reflection
    d = (a*x0 + b*y0 + c) / (a*a + b*b)
    x_ref = x0 - 2*a*d
    y_ref = y0 - 2*b*d
    return x_ref, y_ref

# Example: reflect a triangle across y = x + 1
triangle = [(0,0), (2,0), (1,3)]
m, c = 1, 1

reflected_triangle = [reflect_point(x,y,m,c) for (x,y) in triangle]

# Plot original triangle
x_orig = [p[0] for p in triangle] + [triangle[0][0]]
y_orig = [p[1] for p in triangle] + [triangle[0][1]]
plt.plot(x_orig, y_orig, 'b-', label='Original Triangle')

# Plot reflected triangle
x_ref = [p[0] for p in reflected_triangle] + [reflected_triangle[0][0]]
y_ref = [p[1] for p in reflected_triangle] + [reflected_triangle[0][1]]
plt.plot(x_ref, y_ref, 'r--', label=f'Reflected across y={m}x+{c}')

# Mark vertices
plt.scatter(*zip(*triangle), color='blue')
plt.scatter(*zip(*reflected_triangle), color='red')

plt.legend()
plt.axis('equal')
plt.title("Reflection across y = mx + c")
plt.show()
```

Shearing in x-direction


```python
import matplotlib.pyplot as plt

# Original rectangle vertices (counter-clockwise)
rectangle = [(1,1), (4,1), (4,3), (1,3)]

# Shear factor in x-direction
k = 0.5

def shear_x(point, k):
    x, y = point
    return (x + k*y, y)

# Apply shear to all vertices
sheared_rectangle = [shear_x(p, k) for p in rectangle]

# Close shapes for plotting
x_orig = [p[0] for p in rectangle] + [rectangle[0][0]]
y_orig = [p[1] for p in rectangle] + [rectangle[0][1]]

x_shear = [p[0] for p in sheared_rectangle] + [sheared_rectangle[0][0]]
y_shear = [p[1] for p in sheared_rectangle] + [sheared_rectangle[0][1]]

# Plot both rectangles
plt.plot(x_orig, y_orig, 'b-', label='Original Rectangle')
plt.plot(x_shear, y_shear, 'r--', label='Sheared Rectangle (x-direction)')

# Mark vertices
plt.scatter(*zip(*rectangle), color='blue')
plt.scatter(*zip(*sheared_rectangle), color='red')

plt.legend()
plt.axis('equal')
plt.title("Rectangle Shear in X-direction")
plt.show()
```

Shearing toward both direction

```python
import matplotlib.pyplot as plt

# Original rectangle vertices (counter-clockwise)
rectangle = [(1,1), (4,1), (4,3), (1,3)]

# Shear factors
kx, ky = 0.5, 0.3   # shear in x and y directions

def shear_xy(point, kx, ky):
    x, y = point
    x_new = x + kx * y
    y_new = y + ky * x
    return (x_new, y_new)

# Apply shear to all vertices
sheared_rectangle = [shear_xy(p, kx, ky) for p in rectangle]

# Close shapes for plotting
x_orig = [p[0] for p in rectangle] + [rectangle[0][0]]
y_orig = [p[1] for p in rectangle] + [rectangle[0][1]]

x_shear = [p[0] for p in sheared_rectangle] + [sheared_rectangle[0][0]]
y_shear = [p[1] for p in sheared_rectangle] + [sheared_rectangle[0][1]]

# Plot both rectangles
plt.plot(x_orig, y_orig, 'b-', label='Original Rectangle')
plt.plot(x_shear, y_shear, 'r--', label='Sheared Rectangle (x & y directions)')

# Mark vertices
plt.scatter(*zip(*rectangle), color='blue')
plt.scatter(*zip(*sheared_rectangle), color='red')

plt.legend()
plt.axis('equal')
plt.title("Rectangle Shear in X and Y directions")
plt.show()
```
