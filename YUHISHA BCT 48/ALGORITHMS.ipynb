                  DDA (Digital Differential Analyzer) Line Drawing Algorithm

Algorithm Steps:

1.Start

2.Read the coordinates of the two end points:
    (x1,y1) and (x2,y2)

3.Calculate the differences:
    dx=x2-x1
    dy=y2-y1

4.Determine the number of steps:
    steps= max(∣dx∣,∣dy∣)

5.Calculate the increment values:
    x-increment= dx/steps
    y-increment= dy/steps

6.Set the initial point:
    x=x1, y=y1

7.Plot the point(round(x),round(y))

8.Repeat for steps times:
    x=x+x-increment
    y=y+y-increment
    Plot(round(x),round(y))

9.Stop




                         Bresenham’s Line Drawing Algorithm

Algorithm Steps(for slope <1):

1.Start

2.Read the end points:
    (x1,y1) and (x2,y2)

3.Calculate the differences:
    dx=x2-x1
    dy=y2-y1

4.Initialize the decision parameter:
    p= 2dy-dx

5.Set the initial point:
    x=x1, y=y1

6.Plot the point(x,y)

7.Repeat until x=x2:
   if p<0:
    x=x+1
    p=p+2dy

   else:
    x=x+1
    y=y+1
    p=p+2(dy-dx)
   Plot the point(x,y)

8.Stop
