# 1. Translate a triangle with translation factors tx and ty.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.array([
    [0, 0, 1],
    [2, 0, 1],
    [1, 1, 1]
])

tx = 0.5
ty = 2

m = np.array([
    [1, 0, tx],
    [0, 1, ty],
    [0, 0, 1]
])

tr = m @ t

plt.figure(figsize=(5,5))

plt.plot(
    [t[0,0], t[0,1], t[0,2], t[0,0]],
    [t[1,0], t[1,1], t[1,2], t[1,0]],
    color='black', label='original triangle'
)

plt.plot(
    [tr[0,0], tr[0,1], tr[0,2], tr[0,0]],
    [tr[1,0], tr[1,1], tr[1,2], tr[1,0]],
    color='blue', linestyle='--', label='translated triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("translation of a triangle")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

# 2. Scale a rectangle about an origin.

```python
import numpy as np
import matplotlib.pyplot as plt

r = np.array([
    [1, 4, 4, 1], 
    [1, 1, 3, 3],  
    [1, 1, 1, 1]    
])

sx = 2
sy = 2

s = np.array([
    [sx, 0,  0],
    [0,  sy, 0],
    [0,  0,  1]
])

sr = s @ r

plt.figure(figsize=(5,5))

plt.plot(
    [r[0,0], r[0,1], r[0,2], r[0,3], r[0,0]],
    [r[1,0], r[1,1], r[1,2], r[1,3], r[1,0]],
    color='green', label='original rectangle'
)

plt.plot(
    [sr[0,0], sr[0,1], sr[0,2], sr[0,3], sr[0,0]],
    [sr[1,0], sr[1,1], sr[1,2], sr[1,3], sr[1,0]],
    color='crimson', linestyle='--', label='scaled rectangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scaling of a rectangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

# 3. Rotate a triangle about an origin.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

th = 45
r = np.deg2rad(th)

m = np.array([
    [np.cos(r), -np.sin(r), 0],
    [np.sin(r),  np.cos(r), 0],
    [0,          0,         1]
])

rt = m @ t

plt.figure(figsize=(5,5))

plt.plot(
    [t[0,0], t[0,1], t[0,2], t[0,0]],
    [t[1,0], t[1,1], t[1,2], t[1,0]],
    color='blue', label='original triangle'
)

plt.plot(
    [rt[0,0], rt[0,1], rt[0,2], rt[0,0]],
    [rt[1,0], rt[1,1], rt[1,2], rt[1,0]],
    color='red', linestyle='--', label='rotated triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("rotation of a triangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

# 4. Reflect a triangle about x-axis

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

m = np.array([
    [1, 0, 0],
    [0, -1, 0],
    [0, 0, 1]
])

rt = m @ t

plt.figure(figsize=(5,5))

plt.plot(
    [t[0,0], t[0,1], t[0,2], t[0,0]],
    [t[1,0], t[1,1], t[1,2], t[1,0]],
    color='blue', linestyle='-', label='original triangle'
)

plt.plot(
    [rt[0,0], rt[0,1], rt[0,2], rt[0,0]],
    [rt[1,0], rt[1,1], rt[1,2], rt[1,0]],
    color='black', label='reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about x axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

# 7. Reflect a triangle about y = mx+c

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

m = 1
c = 1
th = np.arctan(m)

T1 = np.array([
    [1, 0, 0],
    [0, 1, -c],
    [0, 0, 1]
])

T2 = np.array([
    [1, 0, 0],
    [0, 1, c],
    [0, 0, 1]
])

R1 = np.array([
    [ np.cos(th),  np.sin(th), 0],
    [-np.sin(th),  np.cos(th), 0],
    [0, 0, 1]
])

R2 = np.array([
    [ np.cos(th), -np.sin(th), 0],
    [ np.sin(th),  np.cos(th), 0],
    [0, 0, 1]
])

Rx = np.array([
    [1, 0, 0],
    [0, -1, 0],
    [0, 0, 1]
])

rf = T2 @ R2 @ Rx @ R1 @ T1
rt = rf @ t

plt.figure(figsize=(6,6))

plt.plot(
    [t[0,0], t[0,1], t[0,2], t[0,0]],
    [t[1,0], t[1,1], t[1,2], t[1,0]],
    color='yellow', linestyle='-', label='original triangle'
)

plt.plot(
    [rt[0,0], rt[0,1], rt[0,2], rt[0,0]],
    [rt[1,0], rt[1,1], rt[1,2], rt[1,0]],
    color='green', linestyle='--', label='reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about y = mx + c")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

# 8. Shear a rectangle about x direction

```python
import numpy as np
import matplotlib.pyplot as plt

r = np.array([
    [1, 4, 4, 1],
    [1, 1, 3, 3],
    [1, 1, 1, 1]
])

shx = 1.5

m = np.array([
    [1, shx, 0],
    [0, 1, 0],
    [0, 0, 1]
])

sr = m @ r

plt.figure(figsize=(5,5))

plt.plot(
    [r[0,0], r[0,1], r[0,2], r[0,3], r[0,0]],
    [r[1,0], r[1,1], r[1,2], r[1,3], r[1,0]],
    color='green', label='original rectangle'
)

plt.plot(
    [sr[0,0], sr[0,1], sr[0,2], sr[0,3], sr[0,0]],
    [sr[1,0], sr[1,1], sr[1,2], sr[1,3], sr[1,0]],
    color='olivedrab', linestyle='--', label='x-direction shear'
)

plt.title("shearing of rectangle in x-direction")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```
