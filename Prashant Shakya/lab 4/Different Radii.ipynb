```python
import matplotlib.pyplot as plt

def draw_ellipse(a, b, cx, cy):
    px = 0
    py = b

    a2 = a * a
    b2 = b * b

    d1 = b2 - a2 * b + 0.25 * a2

    ex, ey = [], []

    while 2 * b2 * px <= 2 * a2 * py:
        ex.extend([cx + px, cx - px, cx + px, cx - px])
        ey.extend([cy + py, cy + py, cy - py, cy - py])

        if d1 < 0:
            px += 1
            d1 += 2 * b2 * px + b2
        else:
            px += 1
            py -= 1
            d1 += 2 * b2 * px - 2 * a2 * py + b2

    d2 = b2 * (px + 0.5) ** 2 + a2 * (py - 1) ** 2 - a2 * b2

    while py >= 0:
        ex.extend([cx + px, cx - px, cx + px, cx - px])
        ey.extend([cy + py, cy + py, cy - py, cy - py])

        if d2 > 0:
            py -= 1
            d2 -= 2 * a2 * py - a2
        else:
            px += 1
            py -= 1
            d2 += 2 * b2 * px - 2 * a2 * py + a2

    return ex, ey


ellipses = [
    (90, 45, 20, -10, 'darkcyan'),
    (55, 25, -120, 80, 'crimson'),
    (35, 65, 140, -70, 'olive')
]

plt.figure(figsize=(7, 7))

for a, b, cx, cy, col in ellipses:
    xs, ys = draw_ellipse(a, b, cx, cy)
    plt.scatter(xs, ys, s=6, color=col, label=f"a={a}, b={b}")

plt.title("Differnet radii")
plt.xlabel("Horizontal")
plt.ylabel("Vertical")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```
