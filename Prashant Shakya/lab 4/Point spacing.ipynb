```python
import matplotlib.pyplot as plt

def midpoint_ellipse_points(rx, ry, center_x, center_y):
    x = 0
    y = ry

    rx_sq = rx * rx
    ry_sq = ry * ry

    decision1 = ry_sq - rx_sq * ry + 0.25 * rx_sq

    region1_x = []
    region1_y = []
    region2_x = []
    region2_y = []

    while 2 * ry_sq * x <= 2 * rx_sq * y:
        region1_x.append(center_x + x)
        region1_y.append(center_y + y)

        if decision1 < 0:
            x += 1
            decision1 += 2 * ry_sq * x + ry_sq
        else:
            x += 1
            y -= 1
            decision1 += 2 * ry_sq * x - 2 * rx_sq * y + ry_sq

    decision2 = ry_sq * (x + 0.5)**2 + rx_sq * (y - 1)**2 - rx_sq * ry_sq

    while y >= 0:
        region2_x.append(center_x + x)
        region2_y.append(center_y + y)

        if decision2 > 0:
            y -= 1
            decision2 += rx_sq - 2 * rx_sq * y
        else:
            y -= 1
            x += 1
            decision2 += 2 * ry_sq * x - 2 * rx_sq * y + rx_sq

    return region1_x, region1_y, region2_x, region2_y

rx, ry = 100, 60
center_x, center_y = 0, 0

r1_x, r1_y, r2_x, r2_y = midpoint_ellipse_points(rx, ry, center_x, center_y)

plt.figure(figsize=(7, 7))
plt.scatter(r1_x, r1_y, color='purple', label='Region 1 (slope < 1)')
plt.scatter(r2_x, r2_y, color='orange', label='Region 2 (slope â‰¥ 1)')

plt.title("Midpoint Ellipse: Point Spacing Comparison")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```
