#### 1. Extend the DDA program to draw a rectangle given two opposite corners.

```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    x_points = []
    y_points = []

    for i in range(int(steps) + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

    plt.plot(x_points, y_points)

x1, y1 = 2, 2
x2, y2 = 8, 6

x3, y3 = x1, y2
x4, y4 = x2, y1

dda(x1, y1, x3, y3)
dda(x3, y3, x2, y2)
dda(x2, y2, x4, y4)
dda(x4, y4, x1, y1)

plt.title("Rectangle using DDA Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

#### 2. Use DDA to draw the axes of a simple coordinate system (X and Y axes).

```python
import matplotlib.pyplot as plt

def dda(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    x_points = []
    y_points = []

    for i in range(int(steps) + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

    plt.plot(x_points, y_points)

dda(-10, 0, 10, 0)

dda(0, -10, 0, 10)

plt.title("Coordinate Axes using DDA Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
