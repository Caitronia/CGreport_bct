# Algorithm: DDA Line Drawing
# step 1: Input the starting and ending co-ordinate of the line:(x1,y1) and (x2,y2)
# step 2: Calculate the differences of the co-ordinate: dx=x2-x1 and dy=y2-y1
# step 3: Determine the number of steps : steps= int(max(abs(dx), abs(dy)))
# step 4: Calculate the increments values: x_inc = dx/steps and y_inc = dy/steps
# step 5: Initialize starting points: x = x1 and y = y1
# step 6: Plot the point(x,y 
# step 7: Repeat the following for steps times:x=x+x_inc and y=y+y_inc,plot the point(round(x),round(y))  
# step 8: End


```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [], []
    
    dx = x2 - x1
    dy = y2 - y1
    
    steps = int(max(abs(dx), abs(dy)))
    
    x_inc = dx / steps
    y_inc = dy / steps
    
    x = x1
    y = y1
    
    for _ in range(steps + 1):
        xes.append(round(x))
        yes.append(round(y))
        x += x_inc
        y += y_inc
        
    return xes, yes

def plot_dda(x1, y1, x2, y2):
    xes, yes = dda_line(x1, y1, x2, y2)
    
    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker='o', linestyle='-')
    plt.title("DDA Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()

plot_dda(2, 6, 24, 14)
```

```python
import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    xes, yes = [], []
    
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    sx = 1 if x2 >= x1 else -1
    sy = 1 if y2 >= y1 else -1
    
    x, y = x1, y1

    if dx >= dy:
        p = 2 * dy - dx
        for _ in range(dx + 1):
            xes.append(x)
            yes.append(y)
            x += sx
            if p >= 0:
                y += sy
                p += 2 * dy - 2 * dx
            else:
                p += 2 * dy
    else:
        p = 2 * dx - dy
        for _ in range(dy + 1):
            xes.append(x)
            yes.append(y)
            y += sy
            if p >= 0:
                x += sx
                p += 2 * dx - 2 * dy
            else:
                p += 2 * dx
    return xes, yes

def plot_bresenham(x1, y1, x2, y2):
    xes, yes = bresenham_line(x1, y1, x2, y2)
    
    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker='o', linestyle='-', color='green')
    plt.title("Bresenham Line Drawing Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()

plot_bresenham(20, 4, 15, 6)
```
