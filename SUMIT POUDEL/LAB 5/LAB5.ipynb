1. Implement Midpoint Ellipse Algorithm

```python
import matplotlib.pyplot as plt

# Function to plot 4 symmetric points
def plot_ellipse_points(xc, yc, x, y, xes, yes):
    points = [
        ( x + xc,  y + yc),
        (-x + xc,  y + yc),
        ( x + xc, -y + yc),
        (-x + xc, -y + yc)
    ]
    for px, py in points:
        xes.append(px)
        yes.append(py)

# Midpoint Ellipse Algorithm
def midpoint_ellipse(rx, ry, xc=0, yc=0):
    x = 0
    y = ry
    rx2 = rx * rx
    ry2 = ry * ry
    xes, yes = [], []

    # Region 1
    p1 = ry2 - (rx2 * ry) + 0.25 * rx2
    plot_ellipse_points(xc, yc, x, y, xes, yes)
    while (2 * ry2 * x) <= (2 * rx2 * y):
        x += 1
        if p1 < 0:
            p1 = p1 + 2 * ry2 * x + ry2
        else:
            y -= 1
            p1 = p1 + 2 * ry2 * x - 2 * rx2 * y + ry2
        plot_ellipse_points(xc, yc, x, y, xes, yes)

    # Region 2
    p2 = (ry2 * (x + 0.5)**2) + (rx2 * (y - 1)**2) - (rx2 * ry2)
    while y >= 0:
        if p2 > 0:
            y -= 1
            p2 = p2 - 2 * rx2 * y + rx2
        else:
            x += 1
            y -= 1
            p2 = p2 + 2 * ry2 * x - 2 * rx2 * y + rx2
        plot_ellipse_points(xc, yc, x, y, xes, yes)

    return xes, yes

# Function to plot the ellipse
def plot_midpoint_ellipse(rx, ry, xc=0, yc=0):
    xes, yes = midpoint_ellipse(rx, ry, xc, yc)
    plt.figure(figsize=(6,6))
    plt.scatter(xes, yes, marker='.', color='purple')
    plt.title("Midpoint Ellipse Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Example: Task 1
plot_midpoint_ellipse(30, 15, 0, 0)
```

2. Draw Ellipses with Different Radii and Centers

```python
# Example: multiple ellipses
plot_midpoint_ellipse(20, 10, 0, 0)     # centered at origin
plot_midpoint_ellipse(25, 15, 10, 10)   # centered at (10, 10)
plot_midpoint_ellipse(15, 30, -10, 5)   # centered at (-10, 5)
```

3. Compare Point Spacing in Region 1 and Region 2

```python
import matplotlib.pyplot as plt

def plot_regions(rx, ry, xc=0, yc=0):
    x = 0
    y = ry
    rx2 = rx * rx
    ry2 = ry * ry
    region1_x, region1_y = [], []
    region2_x, region2_y = [], []

    # Region 1
    p1 = ry2 - (rx2 * ry) + 0.25 * rx2
    while (2 * ry2 * x) <= (2 * rx2 * y):
        region1_x.append(xc + x)
        region1_y.append(yc + y)
        x += 1
        if p1 < 0:
            p1 = p1 + 2 * ry2 * x + ry2
        else:
            y -= 1
            p1 = p1 + 2 * ry2 * x - 2 * rx2 * y + ry2

    # Region 2
    p2 = (ry2 * (x + 0.5)**2) + (rx2 * (y - 1)**2) - (rx2 * ry2)
    while y >= 0:
        region2_x.append(xc + x)
        region2_y.append(yc + y)
        if p2 > 0:
            y -= 1
            p2 = p2 - 2 * rx2 * y + rx2
        else:
            x += 1
            y -= 1
            p2 = p2 + 2 * ry2 * x - 2 * rx2 * y + rx2

    # Plotting both regions with different colors
    plt.figure(figsize=(6,6))
    plt.scatter(region1_x, region1_y, color='blue', label='Region 1')
    plt.scatter(region2_x, region2_y, color='red', label='Region 2')
    plt.title("Midpoint Ellipse: Region 1 vs Region 2")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

# Example: Task 3
plot_regions(30, 15, 0, 0)
```
