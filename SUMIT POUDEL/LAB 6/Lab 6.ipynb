LAB 6: 2D GEOMETRIC TRANSFORMATIONS USING HOMOGENEOUS COORDINATES

Example 1: Rotation by 45 Degrees About Origin


```python
import numpy as np
import matplotlib.pyplot as plt

# Define triangle vertices
triangle = np.array([
    [1, 1, 1],
    [4, 1, 1],
    [2.5, 4, 1]
]).T   # 3x3 matrix (homogeneous form)

# Close the triangle for plotting
triangle_plot = np.hstack((triangle, triangle[:, 0].reshape(3,1)))

# Rotation matrix (45 degrees)
theta = np.pi / 4

R = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta),  np.cos(theta), 0],
    [0, 0, 1]
])

# Apply rotation
rotated = R @ triangle
rotated_plot = np.hstack((rotated, rotated[:, 0].reshape(3,1)))

# Plot
plt.figure()
plt.plot(triangle_plot[0], triangle_plot[1], marker='o', linestyle='-', label="Original Triangle")
plt.plot(rotated_plot[0], rotated_plot[1], marker='s', linestyle='--', label="Rotated Triangle")

plt.title("Rotation of Triangle (45Â°)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

```

Example 2: Scaling (Sx = 2, Sy = 0.5)


```python
import numpy as np
import matplotlib.pyplot as plt

# Scaling matrix
S = np.array([
    [2, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])

scaled = S @ triangle
scaled_plot = np.hstack((scaled, scaled[:, 0].reshape(3,1)))

plt.figure()
plt.plot(triangle_plot[0], triangle_plot[1], marker='o', linestyle='-', label="Original Triangle")
plt.plot(scaled_plot[0], scaled_plot[1], marker='^', linestyle='--', label="Scaled Triangle")

plt.title("Scaling of Triangle")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

Example 3: Translation (tx = 4, ty = 2)


```python
import numpy as np
import matplotlib.pyplot as plt

# Translation matrix
T = np.array([
    [1, 0, 3],
    [0, 1, 2],
    [0, 0, 1]
])

translated = T @ triangle
translated_plot = np.hstack((translated, translated[:, 0].reshape(3,1)))

plt.figure()
plt.plot(triangle_plot[0], triangle_plot[1], marker='o', linestyle='-', label="Original Triangle")
plt.plot(translated_plot[0], translated_plot[1], marker='d', linestyle='--', label="Translated Triangle")

plt.title("Translation of Triangle")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()

```

Lab Task 1
Fixed-point scaling about starting point


```python
import numpy as np
import matplotlib.pyplot as plt

x0, y0 = 2, 3
x1, y1 = 8, 7

x = np.array([x0, x1])
y = np.array([y0, y1])

points = np.vstack([x, y, np.ones_like(x)])

xf, yf = x0, y0

S = np.array([
    [1.5, 0, 0],
    [0, 2, 0],
    [0, 0, 1]
])

T1 = np.array([
    [1, 0, -xf],
    [0, 1, -yf],
    [0, 0, 1]
])

T2 = np.array([
    [1, 0, xf],
    [0, 1, yf],
    [0, 0, 1]
])

M = T2 @ S @ T1
scaled = M @ points

plt.figure()

plt.plot(x, y, marker='o', linestyle='-', label='Original Line')
plt.plot(scaled[0], scaled[1], marker='s', linestyle='--', label='Fixed-Point Scaled')

plt.title("Fixed Point Scaling")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```

Lab Task 2
Scaling about midpoint


```python
import numpy as np
import matplotlib.pyplot as plt

xm = (x0 + x1) / 2
ym = (y0 + y1) / 2

T1 = np.array([
    [1, 0, -xm],
    [0, 1, -ym],
    [0, 0, 1]
])

T2 = np.array([
    [1, 0, xm],
    [0, 1, ym],
    [0, 0, 1]
])

M_mid = T2 @ S @ T1
scaled_mid = M_mid @ points

plt.figure()

plt.plot(x, y, marker='o', linestyle='-', label='Original Line')
plt.plot(scaled_mid[0], scaled_mid[1], marker='d', linestyle='--', label='Midpoint Scaled')

plt.title("Scaling About Midpoint")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
