1. Implementation of Bresenhamâ€™s Line Drawing Algorithm

```python
import matplotlib.pyplot as plt

def bresenham(x1, y1, x2, y2):
    x_points = []
    y_points = []

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    x, y = x1, y1

    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1

    if dx > dy:
        p = 2 * dy - dx
        for i in range(dx + 1):
            x_points.append(x)
            y_points.append(y)
            if p >= 0:
                y += sy
                p += 2 * (dy - dx)
            else:
                p += 2 * dy
            x += sx
    else:
        p = 2 * dx - dy
        for i in range(dy + 1):
            x_points.append(x)
            y_points.append(y)
            if p >= 0:
                x += sx
                p += 2 * (dx - dy)
            else:
                p += 2 * dx
            y += sy

    return x_points, y_points

# Example
x, y = bresenham(2, 3, 15, 9)
plt.plot(x, y, marker='o')
plt.title("Bresenham Line Drawing")
plt.grid(True)
plt.axis('equal')
plt.show()
```

2. Drawing Lines for Different Octants and Visual Comparison with DDA

```python
# DDA Line
def dda(x1, y1, x2, y2):
    x_pts, y_pts = [], []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x, y = x1, y1

    for i in range(steps + 1):
        x_pts.append(round(x))
        y_pts.append(round(y))
        x += x_inc
        y += y_inc

    return x_pts, y_pts

# Compare
x1, y1, x2, y2 = 2, 3, -10, 15

x_d, y_d = dda(x1, y1, x2, y2)
x_b, y_b = bresenham(x1, y1, x2, y2)

plt.plot(x_d, y_d, marker='o', label='DDA')
plt.plot(x_b, y_b, marker='x', label='Bresenham')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.title("DDA vs Bresenham (Different Octant)")
plt.show()
```
