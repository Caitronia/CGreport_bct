## Drawing fundamental shapes

### 1. Points

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_points():
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    
    glPointSize(10)  # size of points
    glBegin(GL_POINTS)
    glColor3f(1, 0, 0)  # Red
    glVertex2f(-0.5, 0.5)
    glColor3f(0, 1, 0)  # Green
    glVertex2f(0.0, 0.5)
    glColor3f(0, 0, 1)  # Blue
    glVertex2f(0.5, 0.5)
    glEnd()
    
    glutSwapBuffers()

glutInit()
glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
glutInitWindowSize(600, 600)
glutInitWindowPosition(100, 100)
glutCreateWindow("OpenGL Points".encode('ascii'))
glutDisplayFunc(draw_points)
glClearColor(0.2, 0.2, 0.2, 1)
glutMainLoop()
```

### 2. Line

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_line():
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    
    glLineWidth(4)
    glBegin(GL_LINES)
    glColor3f(1, 1, 0)  # Yellow
    glVertex2f(-0.7, 0.0)
    glVertex2f(0.7, 0.0)
    glEnd()
    
    glutSwapBuffers()

glutInit()
glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
glutInitWindowSize(600, 600)
glutInitWindowPosition(100, 100)
glutCreateWindow("OpenGL Line".encode('ascii'))
glutDisplayFunc(draw_line)
glClearColor(0.2, 0.2, 0.2, 1)
glutMainLoop()
```

### 3.Triangle

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_triangle():
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    
    glBegin(GL_TRIANGLES)
    glColor3f(1, 0, 1)  # Magenta
    glVertex2f(-0.5, -0.2)
    glVertex2f(0.0, 0.5)
    glVertex2f(0.5, -0.2)
    glEnd()
    
    glutSwapBuffers()

glutInit()
glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
glutInitWindowSize(600, 600)
glutInitWindowPosition(100, 100)
glutCreateWindow("OpenGL Triangle".encode('ascii'))
glutDisplayFunc(draw_triangle)
glClearColor(0.2, 0.2, 0.2, 1)
glutMainLoop()
```

### 4. Quadrilateral

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_quad():
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    
    glBegin(GL_QUADS)
    glColor3f(0, 1, 1)  # Cyan
    glVertex2f(-0.5, 0.5)
    glVertex2f(0.5, 0.5)
    glVertex2f(0.5, -0.5)
    glVertex2f(-0.5, -0.5)
    glEnd()
    
    glutSwapBuffers()

glutInit()
glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
glutInitWindowSize(600, 600)
glutInitWindowPosition(100, 100)
glutCreateWindow("OpenGL Quadrilateral".encode('ascii'))
glutDisplayFunc(draw_quad)
glClearColor(0.2, 0.2, 0.2, 1)
glutMainLoop()
```

### 5. Pentagon

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

def draw_polygon():
    glClear(GL_COLOR_BUFFER_BIT)
    glLoadIdentity()
    
    glBegin(GL_POLYGON)
    glColor3f(1, 0.5, 0)  # Orange
    glVertex2f(0.0, 0.5)
    glVertex2f(0.47, 0.15)
    glVertex2f(0.29, -0.4)
    glVertex2f(-0.29, -0.4)
    glVertex2f(-0.47, 0.15)
    glEnd()
    
    glutSwapBuffers()

glutInit()
glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
glutInitWindowSize(600, 600)
glutInitWindowPosition(100, 100)
glutCreateWindow("OpenGL Pentagon".encode('ascii'))
glutDisplayFunc(draw_polygon)
glClearColor(0.2, 0.2, 0.2, 1)
glutMainLoop()
```

```python
from OpenGL.GL import *
from OpenGL.GLUT import *
from OpenGL.GLU import *

# Window dimensions
width, height = 800, 600

def draw_shapes():
    glClear(GL_COLOR_BUFFER_BIT)  # Clear the screen
    glLoadIdentity()  # Reset transformations

    # --------------------------
    # 1. Draw Points
    # --------------------------
    glPointSize(10)  # size of points
    glBegin(GL_POINTS)
    glColor3f(1.0, 0.0, 0.0)  # Red point
    glVertex2f(-0.8, 0.8)
    glColor3f(0.0, 1.0, 0.0)  # Green point
    glVertex2f(-0.6, 0.8)
    glColor3f(0.0, 0.0, 1.0)  # Blue point
    glVertex2f(-0.4, 0.8)
    glEnd()

    # --------------------------
    # 2. Draw Lines
    # --------------------------
    glLineWidth(3)  # line thickness
    glBegin(GL_LINES)
    glColor3f(1.0, 1.0, 0.0)  # Yellow line
    glVertex2f(-0.8, 0.6)
    glVertex2f(-0.2, 0.6)
    glEnd()

    # --------------------------
    # 3. Draw Triangle
    # --------------------------
    glBegin(GL_TRIANGLES)
    glColor3f(1.0, 0.0, 1.0)  # Magenta triangle
    glVertex2f(-0.7, 0.3)
    glVertex2f(-0.4, 0.6)
    glVertex2f(-0.1, 0.3)
    glEnd()

    # --------------------------
    # 4. Draw Quadrilateral / Polygon
    # --------------------------
    glBegin(GL_QUADS)
    glColor3f(0.0, 1.0, 1.0)  # Cyan quad
    glVertex2f(0.2, 0.5)
    glVertex2f(0.5, 0.5)
    glVertex2f(0.5, 0.2)
    glVertex2f(0.2, 0.2)
    glEnd()

    # --------------------------
    # 5. Draw Polygon (pentagon)
    # --------------------------
    glBegin(GL_POLYGON)
    glColor3f(1.0, 0.5, 0.0)  # Orange polygon
    glVertex2f(-0.5, -0.2)
    glVertex2f(-0.3, -0.2)
    glVertex2f(-0.2, -0.4)
    glVertex2f(-0.4, -0.5)
    glVertex2f(-0.6, -0.4)
    glEnd()

    glutSwapBuffers()  # Swap buffers for double buffering

def main():
    glutInit()
    glutInitDisplayMode(GLUT_RGBA | GLUT_DOUBLE)
    glutInitWindowSize(width, height)
    glutInitWindowPosition(100, 100)
    glutCreateWindow("OpenGL Fundamental Shapes in Python".encode('ascii'))
    glutDisplayFunc(draw_shapes)
    glClearColor(0.2, 0.2, 0.2, 1.0)  # Dark grey background
    glutMainLoop()

if __name__ == "__main__":
    main()
```
