### 1.Translate a Triangle with translation factors tx and ty

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle
triangle = np.array([[1,1], [3,1], [2,3]])

# Translation factors
tx = 2
ty = 2

# Translation
translated = triangle + np.array([tx, ty])

# Plot
plt.figure()
plt.plot(*np.vstack((triangle, triangle[0])).T)
plt.plot(*np.vstack((translated, translated[0])).T)
plt.title("Translation of Triangle")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rectangle = np.array([[1,1], [4,1], [4,3], [1,3]])

scale = 2

scaled = rectangle * scale

plt.figure()
plt.plot(*np.vstack((rectangle, rectangle[0])).T)
plt.plot(*np.vstack((scaled, scaled[0])).T)
plt.title("Scaling of Rectangle About Origin")
plt.show()
```

### 3. Rotation of Triangle About Origin

```python
import numpy as np
import matplotlib.pyplot as plt

triangle = np.array([[1,1], [3,1], [2,3]])

angle = 45
theta = np.radians(angle)

rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta),  np.cos(theta)]])

rotated = triangle.dot(rotation_matrix.T)

plt.figure()
plt.plot(*np.vstack((triangle, triangle[0])).T)
plt.plot(*np.vstack((rotated, rotated[0])).T)
plt.title("Rotation of Triangle (45 degrees)")
plt.show()
```

### d.Reflection about line y = mx + c

```python
import numpy as np
import matplotlib.pyplot as plt

# Original triangle
triangle = np.array([[1,1], [3,1], [2,3]])

# Line parameters y = mx + c
m = 1
c = 1

# Step 1: Translate so line passes through origin
translated = triangle - np.array([0, c])

# Step 2: Rotate line to align with x-axis
theta = np.arctan(m)
rotation_matrix = np.array([[np.cos(-theta), -np.sin(-theta)],
                            [np.sin(-theta),  np.cos(-theta)]])
rotated = translated.dot(rotation_matrix.T)

# Step 3: Reflect about x-axis
reflection_matrix = np.array([[1,0],
                              [0,-1]])
reflected = rotated.dot(reflection_matrix.T)

# Step 4: Reverse rotation
reverse_rotation = np.array([[np.cos(theta), -np.sin(theta)],
                             [np.sin(theta),  np.cos(theta)]])
rotated_back = reflected.dot(reverse_rotation.T)

# Step 5: Reverse translation
final_reflection = rotated_back + np.array([0, c])

# Plot
plt.figure()

plt.plot(*np.vstack((triangle, triangle[0])).T)
plt.plot(*np.vstack((final_reflection, final_reflection[0])).T)

# Draw axes
plt.axhline(0)
plt.axvline(0)

# Draw line y = mx + c
x = np.linspace(-5, 5, 100)
y = m*x + c
plt.plot(x, y)

plt.title("Reflection about line y = mx + c")
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

### 5. Shear of Rectangle
### (a) Shear in X-direction

```python
import numpy as np
import matplotlib.pyplot as plt

rectangle = np.array([[1,1], [4,1], [4,3], [1,3]])

shx = 1

shear_matrix = np.array([[1, shx],
                         [0, 1]])

sheared = rectangle.dot(shear_matrix.T)

plt.figure()
plt.plot(*np.vstack((rectangle, rectangle[0])).T)
plt.plot(*np.vstack((sheared, sheared[0])).T)
plt.title("Shear in X-direction")
plt.show()
```

### (b) Shear in Both Directions

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 4, 4, 1],
    [1, 1, 3, 3],
    [1, 1, 1, 1]
])

# shear factors
shx = 1.0
shy = 0.7

# shearing in both directions
shear_xy = np.array([
    [1,   shx, 0],
    [shy, 1,   0],
    [0,   0,   1]
])

# shearing
sheared_rect = shear_xy @ rectangle

# plot
plt.figure(figsize=(5,5))

plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='dimgray', linestyle='--', label='shearing on both directions'
)

plt.title("shearing of rectangle in both birections")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```
