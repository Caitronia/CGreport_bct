 # DDA (Digital Differential Analyzer) Line Drawing Algorithm

 ### Algorithm Steps:

```python
1. Start

2. Read the coordinates of the two endpoints: (x1,y1) and (x2,y2)

3. Calculate the differences: dx=x2-x1 dy=y2-y1

4. Determine the number of steps: steps max(∣dx∣,∣dy∣)

5. Calculate the increment values: x-increment= dx/steps, y-increment= dy/steps

6.Set the initial point: x=x1, y=y1

7.Plot the point(round(x),round(y))

8.Repeat for steps times: x=x+x-increment y=y+y-increment Plot(round(x),round(y))

9. Stop

                     
```
