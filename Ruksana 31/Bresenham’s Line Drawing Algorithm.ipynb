### Algorithm Steps(for slope <1):

```python
1. Start

2.Read the end points: (x1,y1) and (x2,y2)

3. Calculate the differences: dx=x2-x1 dy=y2-y1

4. Initialize the decision parameter: p 2dy-dx

5.Set the initial point: x=x1, y=y1

6.Plot the point(x,y)

7.Repeat until x=x2: if p<0: x=x+1 p=p+2dy

else: x=x+1 y=y+1 p=p+2(dy-dx) Plot the point(x,y)

8. Stop
```
