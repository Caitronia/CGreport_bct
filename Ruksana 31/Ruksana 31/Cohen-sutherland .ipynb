## 1. Line Completely Inside

```python
import matplotlib.pyplot as plt
x_min, y_min = 40, 40
x_max, y_max = 120, 120

# Line completely inside
x1, y1 = 60, 60
x2, y2 = 100, 100

plt.figure()
plt.title("Completely Inside")

plt.plot([x_min, x_max, x_max, x_min, x_min],
         [y_min, y_min, y_max, y_max, y_min], 'k-')

plt.plot([x1, x2], [y1, y2], 'g--', linewidth=2)

plt.xlim(0, 160)
plt.ylim(0, 160)
plt.grid(True)
plt.show()
```

## 2. Completely outside

```python
import matplotlib.pyplot as plt

x_min, y_min = 40, 40
x_max, y_max = 120, 120

x1, y1 = 10, 60
x2, y2 = 30, 100

plt.figure()
plt.title("Completely Outside")

plt.plot([x_min, x_max, x_max, x_min, x_min],
         [y_min, y_min, y_max, y_max, y_min], 'k-')

plt.plot([x1, x2], [y1, y2], 'g--', linewidth=2)

plt.xlim(0, 160)
plt.ylim(0, 160)
plt.grid(True)
plt.show()
          
```

## 3. Partially Inside (One Intersection)

```python
import matplotlib.pyplot as plt

xmin, ymin = 40, 40
xmax, ymax = 120, 120

x1, y1 = 10, 60
x2, y2 = 100, 60  

plt.plot([xmin, xmax, xmax, xmin, xmin],
         [ymin, ymin, ymax, ymax, ymin], 'k', linewidth=2)

plt.plot([x1, xmin], [y1, y1], 'g--', linewidth=2) 
plt.plot([xmin, x2], [y1, y2], 'p-', linewidth=2)    

plt.xlim(0, 160)
plt.ylim(0, 160)
plt.grid(True)
plt.title("Partially Inside (One Intersection)")
plt.show()
```

## 4.Partially Inside (Two Intersections)

```python
import matplotlib.pyplot as plt

xmin, ymin = 40, 40
xmax, ymax = 120, 120

x1, y1 = 20, 60   # line starts outside left
x2, y2 = 140, 60  # line ends outside right

plt.plot([xmin, xmax, xmax, xmin, xmin],
         [ymin, ymin, ymax, ymax, ymin], 'k', linewidth=2)

plt.plot([x1, xmin], [y1, y1], 'g--', linewidth=2)   
plt.plot([xmin, xmax], [y1, y2], 'p-', linewidth=2)   
plt.plot([xmax, x2], [y2, y2], 'g--', linewidth=2)   

plt.xlim(0, 160)
plt.ylim(0, 160)
plt.grid(True)
plt.title("Partially Inside (Two Intersections)")
plt.show()
```
