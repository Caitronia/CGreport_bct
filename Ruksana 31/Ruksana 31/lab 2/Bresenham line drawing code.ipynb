```python
import matplotlib.pyplot as plt

def bresenham_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1

    p = 2*dy - dx
    x = x1
    y = y1

    xs = [x]
    ys = [y]

    while x < x2:
        if p < 0:
            x = x + 1
            p = p + 2*dy
        else:
            x = x + 1
            y = y + 1
            p = p + 2*(dy - dx)

        xs.append(x)
        ys.append(y)

    print("X points:", xs)
    print("Y points:", ys)

    plt.plot(xs, ys, marker='o')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Bresenham Line Drawing Algorithm")
    plt.grid(True)
    plt.show()

bresenham_line(0, 0, 6, 4)
```
