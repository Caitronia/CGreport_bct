### 1. Implement the DDA algorithm as a function that returns x and y coordinate lists

```python
def dda(x1, y1, x2, y2):
    x_points = []
    y_points = []

    dx = x2 - x1
    dy = y2 - y1

    steps = int(max(abs(dx), abs(dy)))

    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    for i in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc

    return x_points, y_points
```

### 2. Draw lines with different slopes: ùëö < 1, ùëö > 1, horizontal, vertical and negative slope.

```python
import matplotlib.pyplot as plt

lines = [
    (2, 2, 10, 5),    # m < 1
    (2, 2, 5, 10),    # m > 1
    (2, 5, 10, 5),    # horizontal
    (5, 2, 5, 10),    # vertical
    (10, 10, 2, 5)    # negative slope
]

for x1, y1, x2, y2 in lines:
    x, y = dda(x1, y1, x2, y2)
    plt.plot(x, y, marker='o')

plt.title("DDA Line Drawing Algorithm (Different Slopes)")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### 1.Extend the DDA program to draw a rectangle given two opposite corners.

```python
def dda(x1, y1, x2, y2):
    x_points = []
    y_points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for i in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc
    return x_points, y_points
import matplotlib.pyplot as plt
x1 = int(input("Enter x1 (corner 1): "))
y1 = int(input("Enter y1 (corner 1): "))
x2 = int(input("Enter x2 (opposite corner 2): "))
y2 = int(input("Enter y2 (opposite corner 2): "))
x3, y3 = x1, y2
x4, y4 = x2, y1
rect_lines = [
    (x1, y1, x3, y3),  
    (x3, y3, x2, y2),  
    (x2, y2, x4, y4),  
    (x4, y4, x1, y1)   
]

for x_start, y_start, x_end, y_end in rect_lines:
    x, y = dda(x_start, y_start, x_end, y_end)
    plt.plot(x, y, marker='o')

plt.title("Rectangle using DDA Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### 2.Use DDA to draw the axes of a simple coordinate system (X and Y axes).

```python
def dda(x1, y1, x2, y2):
    x_points = []
    y_points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = int(max(abs(dx), abs(dy)))
    x_inc = dx / steps
    y_inc = dy / steps
    x = x1
    y = y1
    for i in range(steps + 1):
        x_points.append(round(x))
        y_points.append(round(y))
        x += x_inc
        y += y_inc
    return x_points, y_points
import matplotlib.pyplot as plt
length = 10
x_axis_x, x_axis_y = dda(-length, 0, length, 0)
y_axis_x, y_axis_y = dda(0, -length, 0, length)
plt.plot(x_axis_x, x_axis_y, color='blue', label='X-axis')
plt.plot(y_axis_x, y_axis_y, color='red', label='Y-axis')
plt.title("Coordinate System using DDA")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```
