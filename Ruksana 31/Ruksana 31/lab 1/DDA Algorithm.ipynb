# LAB-1: DDA (Digital Differential Analyzer)

## Algorithm

**Step 1:** Start  
**Step 2:** Input starting point (x1, y1) and ending point (x2, y2)  
**Step 3:** Compute  
dx = x2 − x1  
dy = y2 − y1  
**Step 4:** steps = max(|dx|, |dy|)  
**Step 5:** x_inc = dx/steps, y_inc = dy/steps  
**Step 6:** Initialize x = x1, y = y1  
**Step 7:** Plot (round(x), round(y))  
**Step 8:** Stop
