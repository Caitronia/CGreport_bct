```python
import matplotlib.pyplot as plt

def draw_line(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))
    x_inc = dx / steps
    y_inc = dy / steps

    x = x1
    y = y1

    xs = [round(x)]
    ys = [round(y)]

    for _ in range(steps):
        x += x_inc
        y += y_inc
        xs.append(round(x))
        ys.append(round(y))

    print("X points:", xs)
    print("Y points:", ys)

    plt.plot(xs, ys)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("DDA Line Drawing Algorithm")
    plt.show()

draw_line(0, 0, 5, 15)
```
