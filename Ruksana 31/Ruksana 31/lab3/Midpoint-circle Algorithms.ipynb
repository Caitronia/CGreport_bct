# Midpoint Circle Algorithm
## Algorithm Steps:

1. **Initialize parameters:**
   - Let the circle radius be `r`.
   - Let the circle center be `(xc, yc)`.
   - Start with the topmost point `(x, y) = (0, r)`.
   - Set the initial decision parameter `p = 1 - r`.

2. **Plot the first set of points:**
   - Use the 8-way symmetry to plot points `(xc±x, yc±y)` and `(xc±y, yc±x)`.

3. **Iterate to calculate the next points:**
   - While `x < y`:
     - Increment `x` by 1.
     - If `p < 0`:
       - `y` remains the same.
       - Update decision parameter: `p = p + 2*x + 1`.
     - Else:
       - Decrement `y` by 1.
       - Update decision parameter: `p = p + 2*(x - y) + 1`.
     - Plot all 8 symmetric points using the new `(x, y)`.

4. **Repeat** until `x >= y`.
