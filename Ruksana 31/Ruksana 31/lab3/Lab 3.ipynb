# LAB-3: Midpoint Circle Algorithm
### 1. Implement the midpoint circle algorithm 


```python
import matplotlib.pyplot as plt

def plot_circle_points(xc,yc,x,y,xes,yes):
    pts =[
        (x+xc, y+yc),
        (-x+xc, y+yc),
        (x+xc, -y+yc),
        (-x+xc, -y+yc),
        (y+xc, x+yc),
        (-y+xc, x+yc),
        (y+xc, -x+yc),
        (-y+xc, -x+yc),
    ]
    for px,py in pts:
        xes.append(px)
        yes.append(py)

def midpoint_circle(r, xc=0, yc=0):
    x=0
    y=r
    p=1-r
    xes, yes =[],[]

    plot_circle_points(xc,yc,x,y,xes,yes)

    while x<y:
        x+=1
        if p<0:
            p= p+2* x+1
        else:
            y-=1
            p= p+2* (x-y)+1
        plot_circle_points(xc,yc,x,y,xes,yes)
    return xes,yes

def plot_midpoint_circle(r, xc=0 , yc=0):
    xes,yes=midpoint_circle(r,xc,yc)
    plt.figure(figsize=(6,6))
    plt.scatter(xes, yes,marker='o',color='red')
    plt.title("Midpoint Circle Algorithm")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.axis('equal')
    plt.show()

plot_midpoint_circle(20,0,0)      
```

### 2. Draw circles with different radii and centres

```python
import matplotlib.pyplot as plt
circles = [
    (10, 0, 0),
    (15, 20, 20),
    (20, -20, -10)
]

plt.figure(figsize=(6,6))

for r, xc, yc in circles:
    xes, yes = midpoint_circle(r, xc, yc)
    plt.scatter(xes, yes) 

plt.gca().set_aspect('equal')  
plt.title("Circles with Different Radii and Centers")
plt.grid(True)
plt.show()
```

### 3. Use the algorithm to draw concentric circles (target pattern).


```python
import matplotlib.pyplot as plt

# Center of all circles
xc, yc = 0, 0

# Radii for concentric circles
radii = [5, 10, 15, 20, 25]

plt.figure(figsize=(6,6))

# Draw each circle
for r in radii:
    xes, yes = midpoint_circle(r, xc, yc)
    plt.scatter(xes, yes)  # Plot circle

plt.gca().set_aspect('equal')  # Keep circles round
plt.title("Concentric Circles (Target Pattern)")
plt.grid(True)
plt.show()
```
