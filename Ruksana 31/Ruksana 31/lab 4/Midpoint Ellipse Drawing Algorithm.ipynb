## Midpoint Ellipse Drawing Algorithm

### Algorithm

1. **Start** the algorithm.

2. **Input** the semi-major axis `a` and semi-minor axis `b` of the ellipse.

3. **Initialize**

   * `x = 0`
   * `y = b`

4. **Calculate the initial decision parameter for Region 1**

   ```
   p1 = b² − a²b + (a² / 4)
   ```

5. **Region 1** (while `2b²x ≤ 2a²y`):

   a. **Plot** the points `(±x, ±y)` using ellipse symmetry.

   b. **If** `p1 < 0`:

   * `x = x + 1`
   * `p1 = p1 + 2b²x + b²`

   c. **Else**:

   * `x = x + 1`
   * `y = y − 1`
   * `p1 = p1 + 2b²x − 2a²y + b²`

6. **Calculate the decision parameter for Region 2**

   ```
   p2 = b²(x + 0.5)² + a²(y − 1)² − a²b²
   ```

7. **Region 2** (while `y ≥ 0`):

   a. **Plot** the points `(±x, ±y)` using ellipse symmetry.

   b. **If** `p2 > 0`:

   * `y = y − 1`
   * `p2 = p2 − 2a²y + a²`

   c. **Else**:

   * `x = x + 1`
   * `y = y − 1`
   * `p2 = p2 + 2b²x − 2a²y + a²`

8. **Stop** the algorithm.
