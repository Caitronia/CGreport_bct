# 1. Translation of a Triangle|

```python
import matplotlib.pyplot as plt
x = [100,150,200,100]
y = [100,50,100,100]
tx = 80
ty = 60
x_new = [xi + tx for xi in x]
y_new = [yi + ty for yi in y]
plt.plot(x,y)          
plt.plot(x_new,y_new)  
plt.title("Translation of Triangle")
plt.show()
```

# 2. Scaling of Rectangle About Origin

```python
import matplotlib.pyplot as plt

x = [50,150,150,50,50]
y = [50,50,120,120,50]

sx = 1.5
sy = 1.5

x_new = [xi * sx for xi in x]
y_new = [yi * sy for yi in y]

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Scaling of Rectangle About Origin")
plt.show()
```

# 3. Rotation of Triangle About Origin

```python
import matplotlib.pyplot as plt
import math

x = [100,150,200,100]
y = [100,50,100,100]

angle = 45
rad = math.radians(angle)

x_new = []
y_new = []

for i in range(len(x)):
    xr = x[i]*math.cos(rad) - y[i]*math.sin(rad)
    yr = x[i]*math.sin(rad) + y[i]*math.cos(rad)
    x_new.append(xr)
    y_new.append(yr)

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Rotation of Triangle")
plt.show()
```

# 4. Reflection of Triangle

```python
import matplotlib.pyplot as plt

x = [100,150,200,100]
y = [100,50,100,100]

x_new = x
y_new = [-yi for yi in y]

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.axhline(0)
plt.title("Reflection About X-axis")
plt.show()
```

```python
x_new = [-xi for xi in x]
y_new = [-yi for yi in y]

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Reflection About Origin")
plt.show()
```

```python
x_new = y
y_new = x

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Reflection About y = x")
plt.show()
```

# 5. Shearing of Rectangle

### (a) Shear in X-direction:

```python
import matplotlib.pyplot as plt

x = [50,150,150,50,50]
y = [50,50,120,120,50]

shx = 0.5

x_new = [x[i] + shx*y[i] for i in range(len(x))]
y_new = y

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Shear in X-direction")
plt.show()
```

### (b) Shear in Both Directions:

```python
shx = 0.5
shy = 0.3

x_new = [x[i] + shx*y[i] for i in range(len(x))]
y_new = [y[i] + shy*x[i] for i in range(len(y))]

plt.plot(x,y)
plt.plot(x_new,y_new)

plt.title("Shear in Both Directions")
plt.show()
```
