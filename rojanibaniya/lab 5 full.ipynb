# 2D GEOMETRIC TRANSFORMATIONS USING HO-MOGENEOUS COORDINATES

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_shape(points, title="Shape"):
    points = np.vstack([points, points[0]])  # Close the shape
    plt.figure()
    plt.plot(points[:, 0], points[:, 1], marker='o')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.gca().set_aspect('equal', 'box')
    plt.grid(True)
    plt.title(title)
    plt.show()


def apply_transformation(points, matrix):
    return (matrix @ points.T).T
```

```python
triangle = np.array([
    [1, 1],
    [3, 1],
    [2, 3]
])

plot_shape(triangle, "Original Triangle")
```

```python
tx, ty = 2, 3

translation_matrix = np.array([
    [1, 0, tx],
    [0, 1, ty],
    [0, 0, 1]
])

triangle_h = np.hstack([triangle, np.ones((3,1))])
translated_triangle = (translation_matrix @ triangle_h.T).T[:, :2]

plot_shape(translated_triangle, "Translated Triangle")
```

```python
rectangle = np.array([
    [1, 1],
    [4, 1],
    [4, 3],
    [1, 3]
])

plot_shape(rectangle, "Original Rectangle")
```

```python
sx, sy = 2, 1.5

scaling_matrix = np.array([
    [sx, 0],
    [0, sy]
])

scaled_rectangle = apply_transformation(rectangle, scaling_matrix)

plot_shape(scaled_rectangle, "Scaled Rectangle")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

# rotation angle (degrees)
theta = 45

# convert degrees to radians
rad = np.deg2rad(theta)

# rotation matrix (about origin)
rotation_mat = np.array([
    [np.cos(rad), -np.sin(rad), 0],
    [np.sin(rad),  np.cos(rad), 0],
    [0,            0,           1]
])

# rotation
rotated_tri = rotation_mat @ triangle

# plot
plt.figure(figsize=(5,5))

# original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='forestgreen', label='original triangle'
)

# rotated triangle
plt.plot(
    [rotated_tri[0,0], rotated_tri[0,1], rotated_tri[0,2], rotated_tri[0,0]],
    [rotated_tri[1,0], rotated_tri[1,1], rotated_tri[1,2], rotated_tri[1,0]],
    color = 'pink', linestyle='--', label='rotated triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("rotation of a triangle about origin")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# triangle coordinates
triangle = np.array([
    [0, 2, 1],   
    [0, 0, 2],  
    [1, 1, 1]   
])

m = 1
c = 1

theta = np.arctan(m)

# translation matrices
T1 = np.array([
    [1, 0, 0],
    [0, 1, -c],
    [0, 0, 1]
])

T2 = np.array([
    [1, 0, 0],
    [0, 1, c],
    [0, 0, 1]
])

# rotation matrices
R1 = np.array([
    [ np.cos(theta),  np.sin(theta), 0],
    [-np.sin(theta),  np.cos(theta), 0],
    [ 0,              0,             1]
])

R2 = np.array([
    [ np.cos(theta), -np.sin(theta), 0],
    [ np.sin(theta),  np.cos(theta), 0],
    [ 0,              0,             1]
])

# reflection about x-axis
Rx = np.array([
    [1,  0, 0],
    [0, -1, 0],
    [0,  0, 1]
])

# combined reflection matrix
reflection_mat = T2 @ R2 @ Rx @ R1 @ T1

# reflection
reflected_tri = reflection_mat @ triangle

#plot
plt.figure(figsize=(6,6))

#original triangle
plt.plot(
    [triangle[0,0], triangle[0,1], triangle[0,2], triangle[0,0]],
    [triangle[1,0], triangle[1,1], triangle[1,2], triangle[1,0]],
    color='black', linestyle='-', label = 'original triangle'
)

#translated triangle
plt.plot(
    [reflected_tri[0,0], reflected_tri[0,1], reflected_tri[0,2], reflected_tri[0,0]],
    [reflected_tri[1,0], reflected_tri[1,1], reflected_tri[1,2], reflected_tri[1,0]],
    color='darkred', linestyle='--', label = 'reflected triangle'
)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("reflection of a triangle about y = mx + c")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 4, 4, 1],
    [1, 1, 3, 3],
    [1, 1, 1, 1]
])

# shear factor in x direction
shx = 1.5

# x-direction shear matrix
shear_x = np.array([
    [1, shx, 0],
    [0, 1,   0],
    [0, 0,   1]
])

# shearing
sheared_rect = shear_x @ rectangle

# plot
plt.figure(figsize=(5,5))

# original rectangle
plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

# sheared rectangle
plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='olivedrab', linestyle='--', label='x-direction shear'
)

plt.title("shearing of rectangle in x direction")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# rectangle coordinates
rectangle = np.array([
    [1, 4, 4, 1],
    [1, 1, 3, 3],
    [1, 1, 1, 1]
])

# shear factors
shx = 1.0
shy = 0.7

# shearing in both directions
shear_xy = np.array([
    [1,   shx, 0],
    [shy, 1,   0],
    [0,   0,   1]
])

# shearing
sheared_rect = shear_xy @ rectangle

# plot
plt.figure(figsize=(5,5))

plt.plot(
    [rectangle[0,0], rectangle[0,1], rectangle[0,2], rectangle[0,3], rectangle[0,0]],
    [rectangle[1,0], rectangle[1,1], rectangle[1,2], rectangle[1,3], rectangle[1,0]],
    color='black', label='original rectangle'
)

plt.plot(
    [sheared_rect[0,0], sheared_rect[0,1], sheared_rect[0,2], sheared_rect[0,3], sheared_rect[0,0]],
    [sheared_rect[1,0], sheared_rect[1,1], sheared_rect[1,2], sheared_rect[1,3], sheared_rect[1,0]],
    color='dimgray', linestyle='--', label='shearing on both directions'
)

plt.title("shearing of rectangle in both birections")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.axis("equal")
plt.show()
```
