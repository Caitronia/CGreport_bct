## lab1 :DDA Algorith

### Alogrithm
### Step 1: Start
### Step 2: Input starting point (x1,y1) and ending point (x2,y2)
### Step 3: Compute
###   dx = x2 − x1
###   dy = y2 − y1
### Step 4: Calculate number of steps
###    steps = max(|dx|, |dy|) 
### Step 5: Compute increment values
### x_inc = dx / steps
    
### y_inc = dy / steps 
### Step 6: Initialize
###  x = x1
  
###  y = y1  
### Step 7: Plot the initial point (x, y)
### Step 8: Repeat for steps times
###   x = x + x_inc
   
###   y = y + y_inc
### Step 9: Plot point (round(x), round(y))
### Step 10: Stop

```python
import matplotlib.pyplot as plt

# Input points
x1, y1 = 2, 2
x2, y2 = 10, 6

dx = x2 - x1
dy = y2 - y1

steps = max(abs(dx), abs(dy))

x_inc = dx / steps
y_inc = dy / steps

x = x1
y = y1

x_points = []
y_points = []

for i in range(int(steps) + 1):
    x_points.append(round(x))
    y_points.append(round(y))
    x += x_inc
    y += y_inc

plt.plot(x_points, y_points, marker='o')
plt.title("DDA Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
